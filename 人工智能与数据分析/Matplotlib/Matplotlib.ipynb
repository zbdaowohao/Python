### Histogram

n, bins, patches = plt.hist(arr, bins=10, normed=0, facecolor='black', edgecolor='black',alpha=1，histtype='bar')

hist的参数非常多，但常用的就这六个，只有第一个是必须的，后面四个可选

arr: 需要计算直方图的一维数组

bins: 直方图的柱数，可选项，默认为10

normed: 是否将得到的直方图向量归一化。默认为0

facecolor: 直方图颜色

edgecolor: 直方图边框颜色

alpha: 透明度

histtype: 直方图类型，‘bar’, ‘barstacked’, ‘step’, ‘stepfilled’

返回值 ：

n: 直方图向量，是否归一化由参数normed设定

bins: 返回各个bin的区间范围

```python
np.random.seed(2019) # 控制每次生成的随机数字保持一致
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000) # 150上下波动

# the histogram of the data n:每个格子的值 ,bins:多少个格子, patches:每个柱状图, facecolor:颜色, alpha:透明度
n, bins, patches = plt.hist(x, 50, normed = 1, facecolor = 'b', alpha = 0.5)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

### Scatter Plot

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2 # 50个半径0-15的圆
plt.scatter(x, y, s = area, c = colors, alpha = 0.5)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('scatter demo')
plt.show()
```

```python
np.random.seed(2019)

dt = 0.001
t = np.arange(0.0, 10.0, dt) #0-10步长0.001
r = np.exp(-t[:1000] / 0.05)
x = np.random.randn(len(t))
s = np.convolve(x, r)[:len(x)] * dt

plt.plot(t, s)
plt.axis([0, 1, 1.1 * np.min(s), 2 * np.max(s)])
plt.xlabel('time (s)')
plt.ylabel('current (nA)')
plt.title('Gaussian colored noise')

a = plt.axes([.65, .6, .2, .2], facecolor = 'k')
n, bins, patches = plt.hist(s, 40, normed = 1)
plt.title('Probability')
plt.xlim(0, 0.2)
plt.xticks([])
plt.yticks([])

a = plt.axes([0.2, 0.6, .2, .2], facecolor = 'k')
plt.plot(t[:len(r)], r)
plt.title('Impulse response')
plt.xlim(0, 0.2)
plt.xticks([])
plt.yticks([])

plt.show()
```

### 3D Plot

```python
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z, rstride = 1, cstride = 1, cmap = cm.viridis)
plt.show()
```
